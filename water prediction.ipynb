```python
X=df1.drop('Potability',axis=1).copy()
Y=df1['Potability'].copy()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
min_scaler=MinMaxScaler()
X=min_scaler.fit_transform(X)
min_X=pd.DataFrame(X)
min_X
```

```python
standard_scaler=StandardScaler()
X=standard_scaler.fit_transform(min_X)
standard_X=pd.DataFrame(X)
standard_X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python

# Step 1: Split the scaled data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(standard_X, Y, test_size=0.2, random_state=42)
```

```python
# Step 2: Initialize the SVM classifier
svm = SVC()
```

```python

# Step 3: Train the SVM classifier using the training data
svm.fit(X_train, Y_train)
```

```python

# Step 4: Make predictions on the testing set
Y_pred = svm.predict(X_test)
```

```python
# Step 5: Evaluate the performance of the SVM classifier
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)
```

```python

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
