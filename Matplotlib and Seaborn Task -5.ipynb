## Qst-1
Import the necessary libraries (Matplotlib and Seaborn) and load the data from the provided CSV file into aPandas DataFrame.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
```

```python
data= pd.read_csv("D:\padas_question.csv")
```

```python
data
```

## Qst-2
Create a scatter plot using Matplotlib to visualize the relationship between height and weight in the dataset. Labelthe axes and give the plot a title.

```python
plt.scatter(x=data['Height (cm)'], y=data['Weight (kg)'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight Scatter Plot')
plt.show()
```

## QST-3
Create a histogram using Matplotlib to visualize the distribution of ages in the dataset. Label the axes and givethe plot a title.

```python
x=data['Age']
plt.hist(x)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

## QST-4
 Create a box plot using Seaborn to visualize the distribution of heights in the dataset by gender. Label the axesand give the plot a title.

```python
sns.boxplot(x=data['Height (cm)'],y=data['Gender'])
plt.xlabel('Height (cm)')
plt.ylabel('Gender')
plt.title('Distribution of heights')
plt.show()
```

## QST-5
Create a bar chart using Seaborn to visualize the average weight of each gender in the dataset. Label the axesand give the plot a title.

```python
sns.barplot(x=data['Gender'],y=data['Weight (kg)'],data=data,ci=None )
plt.xlabel('average Gender')
plt.ylabel('average Weight (kg) by Gender')
plt.title('Average weight of each gender in the dataset')
plt.show()
```
